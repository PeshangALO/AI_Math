# 2.a Coding Problem:
Find the eigenvalues and eigen vectors of the following matrix using Python. 
$$
\left(\begin{array}{cc} 
2 & -1 & 0\\
-1 & 2 & -1\\
0 & -1 & 2
\end{array}\right)
$$ 

```python
import numpy as np
from numpy.linalg import eig

A = np.array([[2,-1,0],[-1,2,-1],[0,-1,2]])
print("A = ", A)

# find eigenvalues and eigenvectors
eigenvalues, eigenvectors = eig(A)
print("eigenvalues = ", eigenvalues)
print("eigenvectors = ", eigenvectors)

```

# 2 b. Coding problem :
Find the eigenvalues and eigen vectors of the following matrix using Python. Verify that the received values indeed satisfy  A.v = l.v

$$
\left(\begin{array}{cc} 
2 & -1 & 0\\
-1 & 2 & -1\\
0 & -1 & 2
\end{array}\right)
$$ 


```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eig

A1 = np.array([[1,2,3],[3,2,1],[1,0,-1]])
eigenvalues, eigenvectors = eig(A1)

eigenvectors_1 = eigenvectors[:,0]
eigenvectors_2 = eigenvectors[:,1]
eigenvectors_3 = eigenvectors[:,2]

# print("Eigenvectors: ", eigenvectors)
print("Eigenvector_1: ", eigenvectors_1)
print("Eigenvector_2: ", eigenvectors_2)
print("Eigenvector_3: ", eigenvectors_3)

eigenvalues_1 = eigenvalues[0]
eigenvalues_2 = eigenvalues[1]
eigenvalues_3 = eigenvalues[2]

print("Eigenvalues: ", eigenvalues_1, eigenvalues_2, eigenvalues_3)
```

 Verify that the received values indeed satisfy  A.v = l.v

```python
# check for the first eigenvector and eigenvalue, 
Av1 = np.dot(A1,eigenvectors_1)
lv1 = eigenvalues_1*eigenvectors_1

# return true if the two vectors are equal
print(np.allclose(Av1, lv1))

# check for the second eigenvector and eigenvalue
Av2 = np.dot(A1,eigenvectors_2)
lv2 = eigenvalues_2*eigenvectors_2

# return true if the two vectors are equal
print(np.allclose(Av2, lv2))

# check for the third eigenvector and eigenvalue
Av3 = np.dot(A1,eigenvectors_3)
lv3 = eigenvalues_3*eigenvectors_3

# return true if the two vectors are equal
print(np.allclose(Av3, lv3))
```

# 3 Coding problem:
  Access the iris dataset (can be found in sklearn.datasets). Calculate the feature-wise mean, the correlation matrix and the variance-covariance matrix. Can you plot the two matrices using seaborn heatmap?


$$
\left(\begin{array}{cc} 
1 & 2 & 3\\
3 & 2 & 1\\
1 & 0 & -1
\end{array}\right)
$$ 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
import pandas as pd

# load the iris dataset
iris = datasets.load_iris()

# create a dataframe of the iris dataset
df = pd.DataFrame(iris.data, columns = iris.feature_names)

# print the first five rows of the dataframe
print(df.head())

# calculate the mean of each feature
print(df.mean())

# calculate the correlation matrix
print(df.corr())

# calculate the variance-covariance matrix
print(df.cov())

# plot the correlation matrix
sns.heatmap(df.corr(), annot=True, cmap='RdYlGn', linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.show()

# plot the variance-covariance matrix
sns.heatmap(df.cov(), annot=True, cmap='RdYlGn', linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.show()

```
