# Task 2

Using the PCA functionality provided by sklearn on Fashion MNIST dataset. 
How many components are required to reach 50% variance?
Can you plot the cumulative variance of 40 components?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml

# Load the data from the OpenML server
X, y = fetch_openml('Fashion-MNIST', version=1, return_X_y=True)

# Normalize the data by dividing by 255
X = X / 255.0
pca = PCA()
pca.fit(X)

```

```python
# Plot the cumulative explained variances
features = range(pca.n_components_)
plt.plot(features, np.cumsum(pca.explained_variance_ratio_), color='black')
plt.xlabel('PCA feature')
plt.ylabel('cumulative variance')
plt.xticks(features)
plt.show()
```

```python
# calculatr how many components are needed to explain 50% of the variance?
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.5) + 1
print(f"{d} explains 50% of the variance")

features = range(10)
plt.plot(features, np.cumsum(pca.explained_variance_ratio_[:10]), color='black')
plt.plot(d, 0.5, "ro")
plt.xlabel('PCA feature')
plt.ylabel('cumulative variance')
plt.xticks(features)
plt.show()
```

```python
# plot the cumulative variance of 40 components
features = range(40)
plt.plot(features, np.cumsum(pca.explained_variance_ratio_[:40]), color='black')
plt.xlabel('PCA feature')
plt.ylabel('cumulative variance')
plt.xticks(features)
plt.show()
```
